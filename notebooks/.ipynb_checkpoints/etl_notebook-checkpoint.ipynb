```python
import numpy as np

df_curated = etl_task.df_raw[
    [
        "Parid",
        "last_sale_date",
        "Owner Name 1",
        "House #",
        "Street Name",
        "Street Suffix",
        "Sale Price",
        "Building Value",
        "Land Value",
    ]
]

for c in ["Street Name", "Street Suffix"]:
    df_curated[c] = df_curated[c].str.title()

for c in ["House #", "Sale Price"]:
    df_curated[c] = df_curated[c].fillna(0).astype(np.int64)

df_curated["Address"] = df_curated[["House #", "Street Suffix", "Street Name"]].apply(
    lambda x: " ".join([x for x in x.dropna().astype(str) if x != "0"]), axis=1
)

df_curated = df_curated.rename(
    columns={
        "Parid": "Parcel ID",
        "last_sale_date": "Sale Date",
        "Owner Name 1": "Owner",
    }
)[["Parcel ID", "Address", "Owner", "Sale Date", "Sale Price", "Building Value", "Land Value"]]

df_curated
```
